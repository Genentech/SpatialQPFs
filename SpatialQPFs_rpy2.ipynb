## Point pattern data analysis
To utilize spatial point process methods, SpatialQPFs provides 3 functions:

- for a single cell type, one can use **Point_pattern_data_uni()**
- for a pair of cell types, one can call **Point_pattern_data_bi()** and **Point_pattern_data_ITLR()**.

For details of the functions, refer to **help(Point_pattern_data_uni)**, **help(Point_pattern_data_bi)**, and **help(Point_pattern_data_ITLR)**

```python
ss_point_uni_lym = SpatialQPFs.Point_pattern_data_uni(path = path, 
                                                      file = file, 
                                                      x_var = "x", 
                                                      y_var = "y", 
                                                      cell_class_var = "cell_id", 
                                                      cell_type = "Lym", 
                                                      scale = 500, 
                                                      myplot = True)
```

```python
print(ss_point_uni_lym)
```

```python
ss_point_uni_TC = SpatialQPFs.Point_pattern_data_uni(path = path, 
                                                     file = file, 
                                                     x_var = "x", 
                                                     y_var = "y", 
                                                     cell_class_var = "cell_id", 
                                                     cell_type = "TC", 
                                                     scale = 500, 
                                                     myplot = True)
```

```python
print(ss_point_uni_TC)
```

## Aggregating spatial feature outputs for downstream analysis

```python
import pandas as pd
import rpy2.rinterface as ri
from rpy2.robjects import pandas2ri

def safe_r_to_pandas(r_object):
    """
    Safely converts an R object from rpy2 to a pandas DataFrame.

    Handles three cases:
    1. The R object is NULL (returns an empty DataFrame).
    2. The R object is already a pandas DataFrame (returns it directly).
    3. The R object is a standard R list/data.frame (converts it).

    Args:
        r_object: The object returned by an rpy2 call.

    Returns:
        A pandas DataFrame.
    """
    # Case 1: The R function returned NULL.
    if isinstance(r_object, ri.NULLType):
        print("Warning: R function returned NULL. Returning an empty DataFrame.")
        return pd.DataFrame()

    # Case 2: The object might have been auto-converted by rpy2 already.
    if isinstance(r_object, pd.DataFrame):
        return r_object

    # Case 3: The object is the expected R list/data.frame with a .names attribute.
    # Check if it has the 'names' attribute before proceeding.
    if hasattr(r_object, 'names'):
        # Your original, correct conversion logic
        df = pd.DataFrame({key: list(r_object.rx2(key)) for key in r_object.names})
        return df
    else:
        print(f"Warning: Received an unexpected R object of type {type(r_object)}. Returning an empty DataFrame.")
        return pd.DataFrame()
```

```python
import pandas as pd
ss_point_uni_TC = safe_r_to_pandas(ss_point_uni_TC)
ss_point_uni_lym = safe_r_to_pandas(ss_point_uni_lym)
ss_point_bi = safe_r_to_pandas(ss_point_bi)
ss_point_ITLR = safe_r_to_pandas(ss_point_ITLR)
ss_lattice_bi = safe_r_to_pandas(ss_lattice_bi)
ss_geostat_bi = safe_r_to_pandas(ss_geostat_bi)
graph_uni_TC = safe_r_to_pandas(graph_uni_TC)
graph_uni_lym = safe_r_to_pandas(graph_uni_lym)
graph_uni_stroma = safe_r_to_pandas(graph_uni_stroma)
graph_all = safe_r_to_pandas(graph_all) 
spat_entropy_all = safe_r_to_pandas(spat_entropy_all)
spat_entropy_TC = safe_r_to_pandas(spat_entropy_TC)
spat_entropy_Lym = safe_r_to_pandas(spat_entropy_Lym)

if not ss_point_uni_TC.empty:
    ss_point_uni_TC.columns = ss_point_uni_TC.columns + '_TC'

if not ss_point_uni_lym.empty:
    ss_point_uni_lym.columns = ss_point_uni_lym.columns + '_Lym'

this_file = pd.DataFrame({"file":[file]})
```

```python
frames = [this_file, ss_point_uni_TC, ss_point_uni_lym, ss_point_bi, ss_point_ITLR, ss_lattice_bi, ss_geostat_bi, graph_uni_TC, graph_uni_lym, graph_uni_stroma,
          graph_all, spat_entropy_all, spat_entropy_TC, spat_entropy_Lym]
this_res = pd.concat(frames, axis=1)
print(this_res)
```
